```python
import matplotlib.pyplot as plt
x = ["Mon", "Tue", "Wed", "Thus","Fri","Sat"]
y = [10,15,5,7,8,1]
plt.plot(x,y)
plt.title('Bakery Sales This Week' )
plt.xlabel('Days')
plt.ylabel('Bakery Sales per day')
plt.show()
```

**matplotlib OOA (object oriented API ) it ll track the current figure behind the scene until u can't call show**



```python

#line plot
#plt.plot(x,y,color = 'color name ' , linestyle = 'Line_Style', linewidth = value , marker = 'marker symbol' , label = 'label name')
import matplotlib.pyplot as plt
months = [1,2,3,4]
sales = [1000,1500,1200,1900]
plt.plot(months, sales , color = 'red',linestyle = '--', linewidth = 2 ,marker = 'o', label = '2025 Sales Data')
plt.title('Monthly Sales Data')
plt.xlabel('Months')
plt.xlim(1,4)
plt.ylim(900,1200)
plt.xticks([1,2,3,4])
plt.yticks([1000,1200,1500,1900],['M1','M2','M3','M4'])
plt.ylabel('Sales per month')
plt.legend(loc = 'upper left')#upper left corner(top left) /bottom right corner(lower right) ,font size
plt.grid(color = 'pink',linestyle = ':',linewidth =1 )
plt.show()
```

```python

#for bar chart
#plt.bar(x,height,color ='color name',width =value,label='label name')
import matplotlib.pyplot as plt
product = ['A','B','C','D']
sales =[1000,1200,800,1500]
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.legend()
plt.bar(product,sales,color = 'red',width = 0.5,label = 'Sales 2025')
plt.show()
```

```python
#pie chart
#shows how each part contribute to the total
#plt.pie(values ,labels = label_list ,colors = color_list,autopct = '%1.1f%%')#value are values to plot(revenue per region),label = slices name ,colors are optional,autopercentage = to write percentage value in piechart(each slice how much hold)
import matplotlib.pyplot as plt
regions = ['North','East','South','West']
revenue = [1000,1200,800,1500]
plt.pie(revenue,labels = regions,autopct = '%1.1f%%', colors=['Gold','Skyblue','coral','lightgreen'])
plt.show()
```

```python
#Histogram
#Group the interval of data
#show frequency in each data , distribution of continuous data by dividing into range (bins)
#plt.hist(data,bins= number_of_bins,color='colorname',edgecolor ='black')

scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]
plt.hist(scores,bins=5,color='blue',edgecolor ='black')#bins -5 part divison(range)
plt.xlabel('Score Range')
plt.ylabel('Frequency')
plt.title('Student Score Distribution')
plt.show()
#use for numerical continous data
```

#scatter plots
is a graph where each point represent one observation with 2 values (x and y)
**syntax**  plt.scatter (x,y,color= 'color_name',marker = 'marker style',label = 'label name' )
X = input data (independent variable) y = output (dependent variable ) , color =dot color , marker = shape of dot (circle, square , triangle ), labels - name for legend



```python
import matplotlib.pyplot as plt
#used to find corelation between 2 variables (examples -advertising budget vs sales , temperature and sales of ice cream )
#use in ml in regression visulisation to visual predicted vs actual value
hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

#scatter plot
plt.scatter(hours_studies,exam_scores,color = 'red',marker = '^',label = 'Exam Scores')
plt.xlabel('Hours studied')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Scores')
plt.grid(True)
plt.legend()
plt.show()
#
```

#Subplots and Layout Adjustments
subplots = 1 canva multiple plots
use for comparison of multiple charts
to quickly analyze related data together
to save space

```python
#used in eda = exploratory data analysis , ml model comparison , weekly sales , Dashboard
#plt.subplot(rows,no of columns,index(1based))
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,25,15,25]

plt.subplot(1,2,1)
#1 row ,2 nd column, 1 position
plt.plot(x,y,color = 'pink')
plt.title('Line chart') # Added title here as it was applied to the second subplot originally
plt.subplot(1,2,2)
#1 row ,2 nd column, 2 position
plt.bar(x,y,color='purple')
plt.title('Bar Chart')
plt.suptitle('Comparison of Line and Bar Charts ')

plt.tight_layout() # Corrected typo from tiget_layout to tight_layout
plt.show()
#
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,25,15,25]

#fig, ax =plt.subplots(no.of rows , no of columns,(optional)figsize= (width,height))
fig,ax = plt.subplots(1,2,figsize = (10,5))
ax[0].plot(x,y)
ax[0].set_title('Line chart')
ax[1].bar(x,y)
ax[1].set_title('Bar Chart')
plt.tight_layout()
plt.show()
```
